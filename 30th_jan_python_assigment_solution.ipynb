4.Check how many times a given number can be divided by 3 before it is less than or equal to 10.

```python
#Answer:
#Programming:

number = int(input("Enter a number: "))
count = 0

while number > 10:
    number = number // 3
    count += 1

print("The number can be divided by 3 : into", count, "times before it is less than or equal to 10.")
```

5.Why and When to Use while Loop in Python give a detalled description with example

Why:
a while loop is used to execute a block of code repeatedly as long as a specified condition is true. The syntax of a while loop in Python is as follows

while condition:
    # code to be executed while the condition is true
When:
The condition is an expression that is evaluated before each iteration of the loop. If the condition is true, then the code inside the loop is executed. Once the code inside the loop has finished executing, the condition is checked again, and if it is still true, the code inside the loop is executed again. This process continues until the condition becomes false.


```python
#Examle:
number = 0
while not (1 <= number <= 10):
    number = int(input("Please enter a number between 1 and 10: "))
print("You entered", number)
```

6.Use nested while loop to print 3 different pattern.

```python
#First Pattern:

i = 1
while i <= 5:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i += 1
```

```python
#2nd Pattern:
i = 5
while i >= 1:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i -= 1
```

```python
#3rd PAttern:
i = 1
while i <= 3:
    j = 1
    while j <= 3 - i:
        print(" ", end="")
        j += 1
    k = 1
    while k <= 2 * i - 1:
        print("*", end="")
        k += 1
    print()
    i += 1
```
